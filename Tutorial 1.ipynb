```python
gather = Semblance().loader()
```

```python
gather
```

```python
shot_gather = gather.data.data.T
shot_gather.shape
```

```python
times = gather.data.twt.values*10**-3 # in seconds
n_samples = biin['Samples'] # no. of samples
dt = gather.attrs['sample_rate']*10**-3 # seconds
```

```python
Semblance.shot_plotter(shot_gather, 2, offset_traces, traces, times)
```

```python
cdps = []
cdps_tr = []
cdps_off = []

for cdp in cdp_no:
    tr = Semblance.cdp_trace(traces-1, cdp_traces-1, cdp-1)
    off = Semblance.cdp_offset(offset_traces, cdp_traces-1, cdp-1)
    cdps_tr.append(tr)
    cdps_off.append(off)
    cdps.append(Semblance.cdp_sort(shot_gather, tr))
```

```python
k = 450  # cdp_no - 1
cdps[k], cdps_tr[k], cdps_off[k]
```

```python
Semblance.cdp_plotter(cdps[k], cdps_tr[k], cdps_off[k], cdp_no[k], times)
```

```python
velocity = Semblance.velocity(1200, 3600+10, 10)
velocity
```

```python
t1 = time.time()
sw = Semblance.semblance_IA(cdps[k], times, cdps_off[k], velocity, window=20, sample_rate=dt)
t2 = time.time()
t2 - t1
```

```python
sw.max(), cdp_no[k]
```

```python
Semblance.semblance_plotter(sw, cdp_no[k], window=20)
```

```python
threshold_value = np.round(sw.max()/2, 2)
th_sw = Semblance.threshold_semblance(sw, threshold_value, cdp_no[k])
```

```python
v_model = Semblance.velocity_model(th_sw, threshold_value, cdp_no[k])
```

```python
source_index, receiver_coordinate = (0, 20, 0), np.array([5.96, 1.2, 0], dtype=float)
```

```python
tt_value, ray_times, ray_velocity = Semblance.eikonal_solver(v_model, source_index, receiver_coordinate)
```

```python
Semblance.travel_time_plotter(v_model, threshold_value, tt_value, source_index, cdp_no[k])
```

```python
Semblance.ray_trace_plotter(v_model, sw, threshold_value, tt_value, ray_times, ray_velocity,
                                 source_index, receiver_coordinate, cdp_no[k], window=20)
```

```python
list_of_percent = [-3, 3, 6]
Semblance.differential_velocity_plotter(sw, ray_velocity, ray_times, list_of_percent, cdp_no[k], window=20)
```

```python
nmo_0 = Semblance.nmo_correction(cdps[k], ray_times, cdps_off[k], ray_velocity, stretch_mute=None)

nmo_m0 = Semblance.nmo_correction(cdps[k], ray_times, cdps_off[k], ray_velocity, stretch_mute=0.3)
```

```python
Semblance.nmo_correct_plotter(cdps[k], nmo_0, nmo_m0, cdp_no[k])
```

```python
nmo_m1 = Semblance.nmo_correction(cdps[k], ray_times, cdps_off[k],
                                       ray_velocity + (list_of_percent[0]/100)*ray_velocity, stretch_mute=0.3)
nmo_m2 = Semblance.nmo_correction(cdps[k], ray_times, cdps_off[k],
                                       ray_velocity + (list_of_percent[1]/100)*ray_velocity, stretch_mute=0.3)
nmo_m3 = Semblance.nmo_correction(cdps[k], ray_times, cdps_off[k],
                                       ray_velocity + (list_of_percent[2]/100)*ray_velocity, stretch_mute=0.3)
```

```python
Semblance.nmo_correct_diff_velocity_plotter(nmo_m0, nmo_m1, nmo_m2, nmo_m3, list_of_percent)
```
